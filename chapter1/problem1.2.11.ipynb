```python
import numpy as np


A = np.array([
    [2/5, 0, 0, 1/5],
    [0, 2/5, 3/10, 1/5],
    [0, 3/10, 2/5, 1/5],
    [3/5, 3/10, 3/10, 2/5]
])
A
```

## Part a

```python
b = np.array([12, 25, 26, 37])
b
```

```python
np.linalg.solve(A, b)
```

## Part b

```python
x = np.array([20, 20, 20, 40])
x
```

```python
np.dot(A, x)
```
